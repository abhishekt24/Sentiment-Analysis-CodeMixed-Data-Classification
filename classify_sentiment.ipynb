```python
import pandas as pd
import numpy as np

# Read in the data
data = pd.read_csv('/home/abhi/Documents/NLP/BE_PRO/english_sentiment/IIITH_Codemixed.csv',sep='\t',engine='python')

# Sample the data to speed up computation
# Comment out this line to match with lecture
#df = df.sample(frac=0.1, random_state=10)

data.head()
```

```python
import random
import time
#random.shuffle(data)
from sklearn.model_selection import train_test_split
#train_data,test_data = train_test_split(data, test_size=0.2, random_state=int(time.time()))
train_data = data[:3500]
test_data = data[3500:]
```

```python
X_train,y_train=train_data['Text'],train_data['Sentiment']
X_test,y_test=test_data['Text'],test_data['Sentiment']
X_train[[0]]
```

```python
import re
REPLACE_NO_SPACE = re.compile("(\.)|(\;)|(\:)|(\!)|(\')|(\?)|(\,)|(\")|(\()|(\))|(\[)|(\])")
REPLACE_WITH_SPACE = re.compile("(<br\s*/><br\s*/>)|(\-)|(\/)")

def preprocess_reviews(reviews):
    reviews = [REPLACE_NO_SPACE.sub("", line.lower()) for line in reviews]
    reviews = [REPLACE_WITH_SPACE.sub(" ", line) for line in reviews]
    
    return reviews

X_train_clean = preprocess_reviews(X_train)
X_test_clean = preprocess_reviews(X_test)
X_train_clean
```

```python
import textblob
from nltk.corpus import stopwords
from textblob import Word
X_train_clean=X_train.apply(lambda x:" ".join(x.lower() for x in x.split()))
X_test_clean=X_test.apply(lambda x:" ".join(x.lower() for x in x.split()))

stop = stopwords.words('english')
X_train_clean = X_train_clean.apply(lambda x: " ".join(x for x in x.split() if x not in stop))
X_test_clean = X_test_clean.apply(lambda x: " ".join(x for x in x.split() if x not in stop))

#X_train_clean= X_train_clean.apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
X_train_clean = X_train_clean.apply(lambda x:''.join([i for i in x if not i.isdigit()]))
#X_test_clean= X_test_clean.apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
X_test_clean = X_test_clean.apply(lambda x:''.join([i for i in x if not i.isdigit()]))
print(X_train_clean[[0]])
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Fit the CountVectorizer to the training data
vect = CountVectorizer().fit(X_train_clean)
vect.get_feature_names()
```

```python
X_train_vectorized = vect.transform(X_train_clean)

X_train_vectorized
```

```python
from sklearn.linear_model import LogisticRegression

# Train the model
model = LogisticRegression()
model.fit(X_train_vectorized, y_train)
```

```python
from sklearn.metrics import roc_auc_score

# Predict the transformed test documents
predictions = model.predict(vect.transform(X_test_clean))
```

```python
import sklearn
sklearn.metrics.confusion_matrix(y_test, predictions, labels=None, sample_weight=None)
```

```python
sklearn.metrics.accuracy_score(y_test, predictions, normalize=True, sample_weight=None)
```
